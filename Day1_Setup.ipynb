```python
from google import genai
client = genai.Client(api_key=os.environ['GEMINI_API_KEY'])

resp = client.models.generate_content(
    model='gemini-2.5-flash',
    contents='Explain the 5-Layer AI Skill Pyramid in 4 short bullet points.'
)
print(resp.text)
```

```python
import os
import getpass
from groq import Groq

# Prompt securely for your Groq API key
os.environ['GROQ_API_KEY'] = getpass.getpass('Paste your Groq API key: ')
```

```python
# Initialize client
client = Groq(api_key=os.environ.get('GROQ_API_KEY'))

# Make the call
chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": "Explain the 5-Layer AI Skill Pyramid in 4 short bullet points.",
        }
    ],
    model="llama-3.3-70b-versatile", # You can check the Groq console for current available models
)

# Print the response
print(chat_completion.choices[0].message.content)
```
